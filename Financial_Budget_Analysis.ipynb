# I can see that it is not easy to understand and analyze the dataset of all years at once, So I decided to create dataframes separately by year.

# Below 5 steps is common for all years 

# 1. Let's create new DataFrame from existing dataset.

```python
df18 =df[['Ministry/Department', 'FY 2018']]
df18
```

# I can see that it is not easy to include all Departments for analysis. For easy understanding I am taking top 10 most funded Ministries/Departments and rest are included in others.

# 2. let’s prepare the data by only selecting the TOP 10 departments and ramaing all the other departments in the other category:

```python
df18.nlargest(10, ['FY 2018'])
```

# 3. Again Create New Dataframe and Insert new row (OTHERS) in the dataframe

```python
df18 = df18.loc[[18,11,7,42,23,14,41,2,0,40],:]
row = {'Ministry/Department':'OTHERS', 'FY 2018':'405424.5'}
df18 = df18.append(row, ignore_index= True)
df18
```

# 4. To see the government's priorities Let's create a bar chart on the above dataframe

```python
fig_18 = px.bar(df18, x='Ministry/Department', y='FY 2018', text_auto='.2s', title="Ministry wise Fund allotted (in ₹crore) for 2018")
fig_18.show()
```

# We can see that the finance department is getting the most of the share from the total budget of the government. 
# 5. Now let’s plot this data into a pie chart to have a clear view of the distribution of funds among all the departments

```python
fig18_pie = px.pie(df18, values='FY 2018', names='Ministry/Department', color='Ministry/Department', color_discrete_sequence=px.colors.sequential.ice, title='Distrubution of fund by % wise for FY 2018')
fig18_pie.show()
```

# We can see that for the FY 2018 finance department is getting 36.6% of the funds and Defence Ministry getting 16.6% of the fund.
